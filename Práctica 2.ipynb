```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Se definen los valores de la señal
n = np.array ([-2, -1, 0, 1, 2, 3]) #Representa las posiciones
x = np.array ([5, 6, 7, 5, 4, 3])  #Representa los valores en las posiciones n

#Para graficar la señal discreta:

plt.stem (n,x)
plt.xlabel("n",color ="green", size= "15")
plt.ylabel("x(n)", color ="green", size= "15")
plt.title("Representación de la Secuencia discreta",color ="orange", size= "20")
plt.grid()
```

```python
# En Python se implementa el impulso unitario mediante la función:

def impulso_unitario(n0,n1,n2):
    n = np.arange (n1,n2+1)
    x = (n == n0).astype(int)
    return x,n

def escalon_unitario(n0,n1,n2):
    n = np.arange(n1, n2+1)
    x = ( n>= n0).astype(int)
    return n,x


def rampa(n0,n1,n2):
    w = np.arange(n1, n2+1)
    z = np.where ( w>= n0, w-n0,0).astype(int)
    return w,z

def compleja(n0,n1,n2):
    w = np.arange(n1, n2+1)
    z = np.where ( w>= n0, w-n0,0).astype(int)
    return w,z

def exponencial(n1,n2,sigma,w0):
    n = np.arange(n1, n2+1)
    p = np.exp (sigma+1j*w0)*n
    return n,p

def sinusoide(n1,n2,w0,theta):
    n = np.arange(n1, n2+1)
    p = np.sin(w0*n+theta)
    return n,p
```

```python
#Para a:

import numpy as np
import matplotlib.pyplot as plt

# Impulso discreto δ(n)
def delta(n):
    return np.where(n == 0, 1, 0)

# Escalón unitario u(n)
def u(n):
    return np.where(n >= 0, 1, 0)

# Rampa r(n)
def r(n):
    return n * u(n)
    
n = np.arange(-4,3)

x1 = 3*delta(n+1) + 5*delta(n+3) + 3*delta(n+2) + 3*delta(n) + delta(n)

plt.stem(n, x1)
plt.title("Secuencia x1(n)",  color = "orange" , size = "20")
plt.xlabel("n", color = "green" , size = "15")
plt.ylabel("Amplitud", color = "green" , size = "15")
plt.grid()
plt.show()
```

```python
#Para b:
n = np.arange(-10,11)

x4 = 2*r(n+3) - r(n-2) - 5*u(n-3)

plt.stem(n, x4)
plt.title("Secuencia x4(n)",  color = "orange" , size = "20")
plt.xlabel("n",  color = "green" , size = "15")
plt.ylabel("Amplitud", color = "green" , size = "15")
plt.grid()
plt.show()
```

6. Genere la siguiente secuencia 
a. Derive la secuencia  
x[𝑛] = { r[n] ,0 ≤𝑛≤5
 𝑟[𝑛 −5], 6 ≤𝑛≤11
𝑟[𝑛 −10]   12 ≤𝑛 ≤17

```python
#Para generar la secuencia por tramos:

n = np.arange(0,18)

x = np.zeros(len(n))

for i in range(len(n)):
    
    if 0 <= n[i] <= 5:
        x[i] = r(n[i])
        
    elif 6 <= n[i] <= 11:
        x[i] = r(n[i]-5)
        
    elif 12 <= n[i] <= 17:
        x[i] = r(n[i]-10)

dx = np.diff(x, prepend=0) #prepend mantiene el tamaño

plt.figure(figsize=(10,8))

plt.subplot(2,1,1)
plt.stem(n,x)
plt.title("Secuencia original", color = "orange" , size = "20"  )
plt.xlabel("n",color ="green", size= "15")
plt.ylabel("x[n]",color ="green", size= "15")
plt.grid()

plt.subplot(2,1,2)
plt.stem(n,dx)
plt.title("Derivada discreta", color = "orange" , size = "20")
plt.xlabel("n",color ="green", size= "15")
plt.ylabel("dx[n]", color ="green", size= "15")
plt.grid()

plt.tight_layout()
plt.show()
```

- Se observa que la señal original es la rampa y esta se repite cada 5 s
- La derivada discreta produce valores constantes (≈1) en las zonas donde la rampa crece.
- En los puntos donde la rampa se reinicia aparecen picos negativos, porque hay un cambio brusco.

La señal derivada no se ve exactamente como la de tiempo continuo porque:

- No existe derivada exacta
- Se usa una aproximación por diferencias lo que produce saltos y otros errores
